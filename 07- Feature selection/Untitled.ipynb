## Regression problem 

```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression

x,y = make_regression(n_samples=50,n_features=5)    # genearate data set using the sklearn lib
```

```python
x = pd.DataFrame(x)
x.head()
```

```python
y[:5]
```

```python
from sklearn.feature_selection import mutual_info_regression    # calcualte the no of mutaual exclusive feature
from sklearn.feature_selection import SelectKBest   # select the best columns 
```

```python
fs = SelectKBest(score_func=mutual_info_regression,k=3)
```

```python
fs.fit(x,y)
```

```python
fs.scores_    # get the most high 3
```

```python
mi_score = pd.Series(fs.scores_,index=x.columns)
mi_score         # we can see the hightes valued  data set so hifhest 3 will tak e
```

```python
x_new  = fs.fit_transform(x,y)
x_new = pd.DataFrame(x_new)
```
