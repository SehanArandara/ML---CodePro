## Types of confustion matrix 

- we can use the confustion matrixes to predict the performance of the data set 

```python
actual = [1,1,1,0,0,1,1,0,0,0,0]
predicted = [1,1,1,0,1,0,1,0,0,1,0]
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(actual,predicted)
```

```python
import pandas as pd
from sklearn.metrics import classification_report
report = pd.DataFrame(classification_report(actual,predicted,output_dict=True))
report
```
