## Random Forest 


- in here the are many trees that is the why we known this as a tree
- so in here we will divide the data set into batches and get the predicted avlues using the Decision Trees
- Finaly we will compare the prediction from trees and get the final prediction 

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../10 - Decision Tree Algorithm/kyphosis.csv')
data.head()
```

```python
x = data.iloc[:,1:4]
x.head()
```

```python
y = data.iloc[:,0]
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test  = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=5)
model.fit(x_train,y_train)
```

```python
pred  = model.predict(x_test)
pred[:5]
```

```python
y_test[0:5]
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,pred)
cm
```
